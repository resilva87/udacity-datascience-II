```python
import pandas as pd
import matplotlib.pyplot as plt


values = {
    'Value': [516,333,296,252,226,184,177,145,145,143],
    'Artist': ["Iron Maiden", "U2", "Metallica", "Led Zeppelin", "Deep Purple", "Lost", "Pearl Jam", "Eric Clapton", "Faith No More", "Lenny Kravitz"]
}

df = pd.DataFrame.from_dict(values)
df = df.sort_values(by=['Value'])
```

```python
%matplotlib inline

ax = df.plot.barh(legend=False, title='Number of Playlists by Artist', grid=True, color='chocolate')
ax.set_yticklabels(df['Artist'])
ax.set_xlabel('# Playlists')
ax.set_ylabel('Artist')
```

```python
import numpy as np
x = values['Value']
x[0] / np.average(x)
```

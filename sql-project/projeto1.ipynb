```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
values = {
    'Value': [3238,
1454,
927,
857,
334,
286,
194,
186,
153,
144],
    'Genre': ["Rock",
"Latin",
"Metal",
"Alternative & Punk",
"Classical",
"Jazz",
"Blues",
"TV Shows",
"R&B/Soul",
"Reggae"]
        }

df = pd.DataFrame.from_dict(values)
df = df.sort_values(by=['Value'])
```

```python

ax = df.plot.barh(legend=False, title='Top 10 Gêneros por Número de Playlists', grid=True, color='chocolate')
ax.set_yticklabels(df['Genre'])
ax.set_xlabel('# Playlists')
ax.set_ylabel('Gênero')
```

```python
x = values['Value']
np.average(x)
```
